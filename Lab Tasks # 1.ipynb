```python
import random

def print_hangman(tries):
    stages = [
        """
           ---------
           |       |
                   |
                   |
                   |
                   |
        ==========
        """,
        """
           ---------
           |       |
           O       |
                   |
                   |
                   |
        ==========
        """,
        """
           ---------
           |       |
           O       |
           |       |
                   |
                   |
        ==========
        """,
        """
           ---------
           |       |
           O       |
          /|       |
                   |
                   |
        ==========
        """,
        """
           ---------
           |       |
           O       |
          /|\      |
                   |
                   |
        ==========
        """,
        """
           ---------
           |       |
           O       |
          /|\      |
          /        |
                   |
        ==========
        """,
        """
           ---------
           |       |
           O       |
          /|\      |
          / \      |
                   |
        ==========
        """
    ]
    return stages[tries]

def get_word():
    words = ["python", "hangman", "programming", "developer", "dynamic", "function", "variable"]
    return random.choice(words).upper()

def hangman():
    word = get_word()
    word_letters = set(word)  # Letters in the word
    guessed_letters = set()  # Correct guesses
    alphabet = set("ABCDEFGHIJKLMNOPQRSTUVWXYZ")

    tries = 0
    max_tries = 6

    print("Welcome to Hangman!")
    print(print_hangman(tries))

    while tries < max_tries and word_letters:
        # Display the word with guessed letters
        current_word = [letter if letter in guessed_letters else "_" for letter in word]
        print("Word:", " ".join(current_word))

        # Get the player's guess
        guess = input("Guess a letter: ").upper()

        if guess in alphabet - guessed_letters:
            guessed_letters.add(guess)
            if guess in word_letters:
                word_letters.remove(guess)
                print(f"Good job! {guess} is in the word.")
            else:
                tries += 1
                print(f"Wrong! {guess} is not in the word.")
        elif guess in guessed_letters:
            print("You already guessed that letter. Try again.")
        else:
            print("Invalid input. Please guess a letter.")

        print(print_hangman(tries))

    if not word_letters:
        print(f"Congratulations! You guessed the word: {word}")
    else:
        print(f"Game Over! The word was: {word}")

if __name__ == "__main__":
    hangman()

```
